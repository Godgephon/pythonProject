```python
import numpy as np
import torch
```

```python
import matplotlib.pyplot as plt
```

```python
net = torch.load('net.pkl')
```

```python
def draw(c, n):
    a = np.ones(n) * c
    b = np.linspace(0, 7, n)
    x = [[a[i], b[i]] for i in range(n)]
    x_tensor = torch.tensor(x, dtype=torch.float)
    out = net(x_tensor)
    y = out.data.numpy()
    plt.plot(b, y, label=str(round(c, 2)))
```

```python
c = np.linspace(0, 1, 10)
```

```python
plt.figure(figsize=(9, 9))
for i in c:
    draw(i, int(1e6))
plt.legend()
plt.grid()
plt.xlabel('input', fontsize=15)
plt.ylabel('output', fontsize=15)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.title('', fontsize=15)
plt.savefig('output.jpg', dpi=100)
```
